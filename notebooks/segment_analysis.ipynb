```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df = pd.read_csv('../data/adobe_segment_financials.csv')

# Display basic structure
print(df.head())
print(df.info())

# Plot segment revenue over time
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='Quarter', y='Revenue ($B)', hue='Segment', marker='o')
plt.title('Adobe Segment Revenue Over Time')
plt.ylabel('Revenue ($B)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
pivot = df.pivot_table(index='Quarter', columns='Segment', values='Revenue ($B)')
pivot_pct = pivot.divide(pivot.sum(axis=1), axis=0) * 100

pivot_pct.plot(kind='bar', stacked=True, figsize=(10, 6), colormap='Set2')
plt.title('Adobe Revenue Mix by Segment (%)')
plt.ylabel('% of Total Revenue')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='Quarter', y='Gross Margin (%)', hue='Segment', marker='o')
plt.title('Adobe Segment Gross Margin Over Time')
plt.ylabel('Gross Margin (%)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
